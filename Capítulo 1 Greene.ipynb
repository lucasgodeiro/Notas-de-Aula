## Conceito de Econometria

Na primeira versão da econometrica, a sociedade de econometria afirma: 

* O principal objetivo da sociedade é promover estudos que objetivam a unificação das abordagens teórico-quantitativo e  empírico-quantitativo para problemas econômicos. 

* Existem vários aspectos da abordagem quantitativa para a economia

* Econometria é a unificação das abordagens estatística, da teoria econômica e matemática.  

## Modelagem Econométrica 

* A modelagem econométrica começará geralmente com uma proposição teórica. 

* Por exemplo, a função consumo de Keynes, onde a propensão marginal a consumir de depende do nível de renda $X$, ou seja $C=f(X)$.  

* Sabemos que propensão marginal a consumir $dC/dX$ é positiva e menor que a unidade. 

* Também sabemos que a propensão média a consumir $\frac{X}{C}$ cai à medida que a renda aumenta. 

* A função consumo mais comum pode ser escrita de forma linear:
$$
C = \alpha + \beta X
$$
 
* Usando a econometria e um conjunto de dados poderíamos testar as hipóteses da teoria, ou seja, $\alpha >0 $ e $0< \beta < 1 $.  

* Poderíamos testar se os parâmetros são constantes ao longo do tempo ou se mudam de acordo com as gerações, por exemplo. 

* Se existe diferenças sistemáticas entre os países. 

* Se existem outros fatores que melhorem a habilidade da renda explicar o consumo.
