# O Modelo Clássico de Regressão Linear Múltipla 

## Introdução 

* Um estudo econométrico começa com um conjunto de proposições sobre aspectos da economia. Exemplos familiares são equações de demanda, funções de produção e modelos macroeconômicos. 

* A investigação empírica fornece estimações de parâmetros desconhecidos no modelo, tal como elasticidades ou os efeitos da política monetária, e usualmente tentam medir a validade da teoria contra o comportamento dos dados observados. 

* Uma vez que o modelo adequado é construído, ele pode ser usado para previsões dentro e fora da amostra ou para a análise do comportamento. 

* A **Regressão Linear** é a ferramenta mais útil do kit dos econometristas. 

## O Modelo de Regressão Linear

 O **Modelo de Regressão Linear Múltipla** é usado para estudar a relação entre uma **Variável Dependente** e uma ou mais **Variáves Independentes**. A forma genérica do modelo de regressão é: 
 
 \begin{align} 
 y = f(x_{1},x_{2}, \ldots, x_{K} ) + \varepsilon \\
   y = x_{1}\beta_{1} + x_{2}\beta_{2}, \ldots , x_{K}\beta_{K} + \varepsilon 
 \end{align} 

Onde $y$ é a variável dependente ou **explicada** e $x_{1}, \ldots, x_{k} $ são as variáveis independentes ou **explicativas**. A teoria especificará $f(x_{1},x_{2}, \ldots , x_{K})$. 

Esta função é comumente chamada de **equação de regressão da população** de $y$ sobre $x_{1}, \ldots, x_{K}$. Nesta configuração, $y$ é o **regressando** e $x_{k}$, $k = 1, \ldots, K$ são os **regressores** ou **co-variáveis**. 

A teoria subjacente especificará quais serão as variáveis dependentes e independentes. 

O termo $\varepsilon$ é a perturbação aleatória ou resíduo. O resíduo surge por vários motivos:

* Não conseguimos captar todas as influências sobre as variáveis econômicas no modelo.

* O efeito líquido, seja ele positivo ou negativo, dos fatores omitidos é capturado pelas perturbações do modelo. 

* Outro questães capturados pelos resíduos são os erros de medida. Por exemplo, a dificuldade de se obter medidas razoáveis de lucros, taxas de juros, estoque de capital ou fluxos de serviços do estoque de capital é um tema recorrente na literatura empírica. 

* No caso extremo, podemos ter uma variável do modelo teórico não observável, como no modelo de renda permanente do consumo [Friedman(1957)]. 

Agora, vamos assumir que cada observação na amostra $(y_{i},x_{i1},x_{i2}, \ldots , x_{iK} )$, $i = 1, \ldots , n$, é gerada pelo processo descrito por: 

\begin{align} 
y_{i} = x_{i1} \beta_{1} + x_{i2} \beta_{2} + \ldots + x_{iK} \beta_{K} + \varepsilon_{i} 
\end{align} 
 
Os valores observados de $y_{i}$ é a soma de duas partes, uma determinística e uma parte aleatória $\varepsilon_{i}$. Nosso objetivo é estimar os parâmetros desconhecidos do modelo, usar os dados para estudar a validade da proposição teórica, e possivelmente usar o modelo para prever a variável $y$.

### Exemplo 2.2 

Uma das vantagens da regressão mùltipla é identificar a efeitos independentes de um conjunto de variáveis sobre a variável dependente. 

Estudos recentes tem analisado a relação entre rendimento e educação. Esperamos que, na média, maiores níveis educacionais são associados com rendas mais altas. O modelo de regressão simples: 

$$
earnings = \beta_{1} + \beta_{2}education + \varepsilon,
$$

entretanto, negligencia o fato de que a maioria das pessoas tem rendas mais altas quando estão mais velhos do que quando estão novos. Assim, $\beta_{2}$ irá superestimar o efeito marginal do impacto da educação. 

Se idade e educação são positivamente correlacionados, então o modelo de regressão associará todos os aumentos observados na renda com aumento na educação. Uma melhor especificação levaria em conta os efeitos da idade: 

$$
earnings = \beta_{1} + \beta_{2}education +  + \beta_{3}age + \varepsilon,
$$

Também é frequentemente observado que a renda tende a aumentar menos rapidamente nos ùltimos anos de rendimento do que nos mais recentes. Para acomodar essa possibilidade, podemos extender o modelo para: 


$$
earnings = \beta_{1} + \beta_{2}education +  + \beta_{3}age + \beta_{4}age^{2} + \varepsilon,
$$

Assim, esperaríamos $\beta_{3}$ positivo e $\beta_{4}$ negativo. 

Com este modelo poderíamos comparar rendas de dois indivíduos com a mesma idade e níveis educacionais diferentes. 

Para estudarmos o impacto da educação, teríamos que observar o indivíduo duas vezes, o que na prática não é possível. 

Se consideramos educação como um tratamento, então o objetivo real do experimento é medir o impacto do tratamento sobre o tratado. O capítulo 19 tratará com mais detalhes sobre o assunto. 




## Hipóteses do Modelo de Regressão Linear

1. **Linearidade:** $y_{i} = x_{i1}\beta_{1} + x_{i2}\beta_{2} + \cdots x_{iK}\beta_{K} + \varepsilon_{i} $. O modelo especifica uma relação linear entre $y$ e $x_{1}, \cdots, x_{K}$. 

2. **Posto completo:** Não existe uma relação linear exata entre quaisquers variáveis independentes. Esta hipótese será necessária para a estimação dos parâmetros do modelo.

3. **Exogeneidade das variáveis independentes:** $E[\varepsilon{i} | x_{j1} , x_{j2}, \cdots , x_{jK}] = 0 $. Ou seja, o valor esperado do resíduo na observação $i$ na amostra não é uma função das variáveis independentes observadas em qualquer observação, incluindo estas. Isto significa que as variáveis independentes não carregam informação útil para a previsão de $\varepsilon_{i}$. 

4. **Homocedasticidade e não autocorrelação:** Cada perturbação $\varepsilon_{i}$ tem a mesma variância finita, $\sigma^{2}$, e é não correlacionada com qualquer outra perturbação $\varepsilon_{j}$. Esta hipótese limita a generalidade do modelo, e examinaremos como relaxá-la nos capítulos segunites. 

5. **Geração dos dados:** Os dados em $x_{j1}, x_{j2}, \cdots, x_{jK} $ podem ser uma mistura de constantes e variáveis aleatórias. O elemento crucial para os presentes objetivos são a hipótese 3 de estrita independência das médias e a hipótese de independência da variância implícita em 4. As análises serão feitas condicionalmente sobre o $\mathbf{X}$ observado, então se os elementos em $\mathbf{X}$ são constanttes fixas ou gerações aleatórias de um processo estocástico que não influenciará os resultados. 

6. **Distribuição normal:** As perturbações são normalmente distribuídas. 


### Hipótese 1: Linearidade do Modelo de Regressão 

Seja $\mathbf{x}_{k} $ as $n$ observações sobre a variável $x_{k}$, $k=1, \ldots, K$ e organizando os dados em uma matrix de dados $\mathbf{X} $, $ n \times K$. Na maioria dos contextos a primeira coluna de $ \mathbf{X} $ é uma coluna com $1s$ a fim de que possamos estimar o termo constante do modelo $\beta_{1}$.  Seja $\mathbf{y}$ as $n$ observaçõs $y_{1}, y_{2}, \ldots y_{n}$ e seja $\mathbf{\varepsilon}$ um vetor coluna contendo as $n$ perturbações do modelo.   

O modelo pode ser escrito como: 

\begin{equation}
\mathbf{y} = \mathbf{x}_{1}\beta_{1} + \cdots + \mathbf{x}_{K}\beta_{K} + \mathbf{\varepsilon}
\end{equation} 

Na forma da hipótese 1, podemos escrever: 

\begin{equation} 
\mathbf{y} = \mathbf{X} \mathbf{\beta} + \mathbf{\varepsilon}.
\end{equation}

Para o modelo de regressão ser linear, podemos usar ambos, as variáveis originais ou alguma transformação adequada, como por exemplo, o modelo

$$
y = Ax^{\beta} e^{\varepsilon}
$$

é linear depois de aplicarmos o log em ambos os lados da equação, ao passo que o modelo: 

$$
y = Ax^{\beta} + \varepsilon
$$

não é linear. A variá dependente observada é a soma de dois componentes, um elemento determinístico $\alpha + \beta$ e uma variável aleatória $\varepsilon $. Vale a penas destacar que nenhuma das duas partes é diretamente observada, porque $\alpha$ e $\beta$ são desconhecidos.  

No contexto de regressão linear, a linearidade se refere a forma como os parâmetros e as perturbações entram na equação, não necessariamente a relação entre as variáveis. Por exemplo, as equações $y = \alpha + \beta x + \varepsilon $, $y = \alpha + \beta \cos(x) + \varepsilon $  e $y = \alpha + \beta/x + \varepsilon $ são todas lineares em alguma função de $x$ pelo definição que usamos aqui. Nos apenas $x$ foi transformado, mas $y$ também poderia ser, com em $y = Ax^{\beta}e^{\varepsilon} $, na qual é uma relação linear nos logs de $x$ e $y$; $\ln y = \alpha + \beta \ln x + \varepsilon $.  Este tipo de modelo é usado em várias formas funcionais, como por exemplo o modelo **loglinear**: 

\begin{equation} 
\ln y = \beta_{1} + \beta_{2} \ln x_{2} + \beta_{3} \ln x_{3} + \cdots + \beta_{K} \ln x_{K} + \varepsilon. 
\end{equation} 

Esta equação também é conhecida como elasticidade constante, pois a elasticidade de $y$ em relação a $x$ é $\partial \ln y / \partial \ln x_{k} = \beta_{k} $ no qual não varia com $x_{k}$. A forma loglinear é bastante usada em modelo de produção e demanda. 





### Exemplo 2.3 O Mercado de Gasolina nos Estados Unidos

Os dados sobre o mercado de gasolina nos estados unidos estão disponíveis para os anos de 1953-2004 e são dados na tabala F2.2 no apêndice F. Usaremos estes dados para obter, entre outras coisas, estimativas da renda, do próprio preço e elasticidades cruzadas da demanda para este mercado. Estes dados também apresentam uma interessante questão sobre o ponto controverso de manter "todas as outras coisas constantes", que foi sugerido. Em particular, considere um modelo um tanto abreviado do consumo de gasolina per capita: 

$$
\ln (G/pop) = \beta_{1} + \beta_{2} \ln (Income/pop) + \beta_{3} \ln price_{G} + \beta_{4} \ln P_{newcars} + \beta_{5} \ln P_{usedcars} + \varepsilon.
$$

Este modelo nos trará estimativas da elasticidades da demanda e do preço por gasolina e uma estimativa da elasticidade demanda com respeito ao preços de carros novos e usados. 

O que devemos esperamos para o sinal de $\beta_{4}$? Carros e gasolina são bens complementares, então se os preços dos carros novos sobem, ceteris paribus, o consumo de gasolina deve cair. Deveria? Se os preços dos carros novos aumentam, então os consumidores comprarão menos; eles manterão seus carros usados por mais tempo e comprarão menos novos. 

Se os carros usados usam mais gasolina do que os novos, então o aumento dos preços dos carros novos levariam a um aumento no consumo de gasolina, não mais baixo. Podemos usar o modelo de regressão múltipla e os dados da gasolina para tentar responder essa questão. 


```r
# Exemplo 2.3 
## Baixando os dados 
f23 <- read.csv('http://www.stern.nyu.edu/~wgreene/Text/Edition7/TableF2-2.csv',header = TRUE)
head(f23)
```

```r
# Obtendo as variáveis do modelo 
## Consumo de gasolina
G <- f23$GASEXP
## Preço da gasolina 
P_G <- f23$GASP
## Renda 
INCOME <- f23$INCOME
## Preços dos carros novos e usados
P_NC <- f23$PNC
P_UC <- f23$PUC
## Populaçao 
POP <- f23$POP
## Anos
YEAR <- f23$YEAR
```

```r
## Agora vamos estimar o modelo de regressão linear, usando a função lm
gas_model <- lm(log(G/POP) ~ log(P_G) + log(INCOME/POP) + log(P_NC) + log(P_UC))
summary(gas_model)
```

### Modelo Semilog 

O **modelo semilog** é frequentemente usado para taxas de crescimento: 

$$
\ln y = \mathbf{ x^{\prime}_{t} \beta} + \delta t + \varepsilon_{t}
$$

Neste modelo, o termo a proporção autônoma(pelo menos não explicado pelo próprio modelo), ou seja, a taxa de crescimento do período é $\partial \ln y/ \partial t = \delta $. Outras variações da forma geral pode ser descritas como: 

$$
f(y_{t}) = g(\mathbf{ x^{\prime}_{t} \beta} + \varepsilon_{t}  )
$$

permitirá uma variedade tremenda de formas funcionais, na qual todas se ajustarão a nossa definição de modelo linear. 


### Hipótese 2: Posto Completo

A hipótese 2 é que não existe uma relação linear exata entre as variáveis. 


Hipótese: $\mathbf{X}$ é uma matrix $ n \times K$ com posto $K$. 

Assim, $\mathbf{X}$ tem posto completo; as colunas de $\mathbf{X}$ são linearmente independentes e existem no mínimo $K$ observações. Essa hipótese é conhecida como a condição de identificação. 

Relembrando que o posto de uma matriz é o número de linhas não nulas na sua forma escalonada.  Nas matrizes com posto completo, um sistema de equação linear terá solução única. 

Exemplo, resolva o seguinte sistema de equação linear usando o escalonamento de matrizes: 


\begin{eqnarray} 
4 x_{1} + 2 x_{2} = 6 \\
2 x_{1} + 2 x_{2} = 4
\end{eqnarray} 

A matriz tem posto completo? O sistema tem solução única? 

Agora resolva o seguinte sistema usando escalonamento: 

\begin{align} 
4 x_{1} + 2 x_{2} = 6 \\
8 x_{1} + 4 x_{2} = 12
\end{align} 

A matriz tem posto completo? O sistema tem solução única? 





### Exemplo 2.5 

Suponha que o seguinte modelo cross section relaciona o consumo $C$ a renda: 

$$
C = \beta_{1} + \beta_{2} nonlabor\,\,\,income + \beta_{3} salary + \beta_{4} total\,\,\,income + \varepsilon
$$

onde a renda total(total income) é claramente igual ao salário(salary) mais a renda não proveniente do trabalho(nonlabor income). Assim, temos uma relação clara de dependência linear entre as variáveis. 

Nesta situação, não temos como estimar os parâmetros do modelo. 

Se existem menos que $K$ observações, então $\mathbf{X}$ não pode ter posto completo. Assim, fazemos a hipótese redundante de que $n$ tem que ter pelo menos a dimensão de $K$, ou seja, para MQO $K>n$ não é possível de ser estimado.



## Hipótese 3:  Exogeneidade das Variáveis Independentes 

Assume-se que o resíduo tem valor esperado condicional zero para cada observação, no qual podemos escrever: 

$$
E[\varepsilon_{i} | \mathbf{X} ] = 0
$$

Para o conjunto completo de observaçõs, podemos escrver a hipótese 3 como: 

$$
E[\mathbf{\varepsilon} | \mathbf{X} ] = 
\begin{bmatrix} 
E[\varepsilon_{1} | \mathbf{X} ] = 0 \\
E[\varepsilon_{2} | \mathbf{X} ] = 0 \\
\vdots \\ 
E[\varepsilon_{n} | \mathbf{X} ] = 0
\end{bmatrix} 
= 
\mathbf{0} 
$$

O lado esquerdo da equação afirma que a média de cada resíduo condicionado sobre todas a observações de $\mathbf{x}_{i} $ é zero. 

A média condicional significa, em outras palavras que nenhuma observação de $\mathbf{x} $ carrega informação sobre o valor esperado das perturbações. 

Em resumo, até este ponto , assumimos que as perturbações são puramente sorteios aleatórios de alguma população. 

A média condicional zero implica que o média não condicional também é zero, desde que: 

$$
E[\varepsilon_{i}] = E_{\mathbf{x}}[E[\varepsilon_{i} | \mathbf{X} ]     ] = E_{ \mathbf{x} }[0] = 0 
$$

Desde que , para cada $\varepsilon_{i}$ , $ Cov[ E[\varepsilon_{i} | \mathbf{X}  ],\mathbf{X}   ] = Cov [\varepsilon_{i} | \mathbf{X} ] $, a hipótese 3 implica que $Cov[\varepsilon_{i} | \mathbf{X} ]  = 0  $  para todo $i$. 

A hipótese 3 implica também que: 

$$
E[\mathbf{y} | \mathbf{X}  ] = \mathbf{X} \mathbf{\beta}  
$$

Ou seja, a regressão de $\mathbf{y} $ sobre $\mathbf{X}$ é a média condicional $E[\mathbf{y} | \mathbf{X}  ]  $.
